```python
fruits = ["apple", "mango", "guava", "strawberry", "banana"]

print("\nFinding elements in List: \n")
print(f" fruits[0]: '{fruits[0]}'")
print(f" fruits[-1]: '{fruits[-1]}")
print(f" fruits[1:4]: '{fruits[1:4]}' ")
```

```python
print("Mutability")
fruits[0] = "apricot"
print(f"After fruits[0] = apricot: {fruits}")
```

```python
# adding elements
print(f"\n Adding elements: \n")
fruits.append("fig")
print(f"List after appending: {fruits}")
```

```python
fruits.insert(1, "blueberry")
print(f"New element added at index 1\n")
print(f"List after insert operation: {fruits}")
```

```python
fruits.pop(1)
```

```python
more_fruits = ["orange", "raspberry"]
fruits.extend(more_fruits)
print(f"List after extend: {fruits}")
```

```python
fruits.remove("fig")
print(f"List after removal: {fruits}")
```
