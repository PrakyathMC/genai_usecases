# While

```python
age = 5
while age <= 8:
    print("The while loop in python")
    age += 1
    if(age == 7):
        break
```

```python
countries = ("India", "USA", "UK", "Canada")
for i in range(len(countries)):
    if countries[i] == "UK":
        print("This is", countries[i])
        break
else:
    print("it got here")
```

```python
countries = ("India", "USA", "UK", "Canada")
for i in range(len(countries)):
    if countries[i] == "USA":
        print("This is", countries[i])
        continue
    print("it got here")
```
